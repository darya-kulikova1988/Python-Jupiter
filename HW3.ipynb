```python
#1.0 Считать данные с помощью pandas
#Вывести на экран первые 5 строк

import pandas as pd
import numpy as np

df = pd.read_csv('kc_house_data.csv')
df.head()
# df.info()
```

```python
#1.1 Создать новый признак price_per_sq_lot, который будет содержать 
#среднюю стоимость за один кв. метр общей площади

df['price_per_sq_lot']=df['price']/df['sqft_lot']

df.head()
```

```python
#1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между
# годом реновацией дома и годом постройки дома. Если реновации дома не было, то в новом признаке поставьте 0

def delta_renovated(row):
    if row['yr_renovated']==0:
        return 0
    else:
        return row['yr_renovated']-row['yr_built']

df['delta_renovated']=df.apply(delta_renovated,axis=1)

df.head()
```

```python
#1.3 Создайте признаки года продажи, месяца продажи

df['date']=pd.to_datetime(df['date'],format='%Y%m%dT%H%M%S')
df['year_of_sale']=df['date'].dt.year
df['month_of_sale']=df['date'].dt.month_name()

df.head()
```

```python
#1.4 Удалите признаки date, zipcode, lat, long
df.drop(columns=['date', 'zipcode', 'lat', 'long'],inplace=True)

df.head()
```
